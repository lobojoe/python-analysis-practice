```python
def load_list_data(lst, csv_file, column_name):
    with open(csv_file) as csv_info:
        csv_dict = csv.DictReader(csv_info)
        for row in csv_dict:
            lst.append(row[column_name])
        return lst
```

```python
# look at the data in insurance_csv_dict
load_list_data(ages, insurance, 'age')
load_list_data(sexes, insurance, 'sex')
load_list_data(bmis, insurance, 'bmi')
load_list_data(num_children, insurance, 'children')
load_list_data(smoker_statuses, insurance, 'smoker')
load_list_data(regions, insurance, 'region')
load_list_data(insurance_charges, insurance, 'charges')
```

```python
class Patients_Info:
    def __init__(self, patients_ages, patients_sexes, patients_bmis, patients_num_children, patients_smoker_statuses, patients_regions, patients_insurance_charges):
        self.patients_ages = patients_ages
        self.patients_sexes = patients_sexes
        self.patients_bmis = patients_bmis
        self.patients_num_children = patients_num_children
        self.patients_smoker_statuses = patients_smoker_statuses
        self.patients_regions = patients_regions
        self.patients_insurance_charges = patients_insurance_charges
    
    def average_age(self):
        total_age = 0
        for age in self.patients_ages:
            # sum of the total age
            total_age += int(age)
        return 'Average age: ' + str(round(total_age / len(self.patients_ages), 2)) + ' years old.'
    
    def average_charge(self):
        total_charge = 0
        for charge in self.patients_insurance_charges:
            total_charge += int(float(charge))
        return 'Average charge ' + str(round(total_charge / len(self.patients_ages), 2)) + ' dollars.'
    
```

```python
patient_info = Patients_Info(ages, sexes, bmis, num_children, smoker_statuses, regions, insurance_charges)
```

```python
def sw_charges(selected_column):
    sw_total = 0
    sw_charge = 0
    for row in master_insurance.values():
        if row['Region'] == 'southwest':              
            sw_total += 1
            sw_charge += row[selected_column]
    return 'Average charge for the southwest region: ' + str(round(sw_charge / sw_total, 2)) + ' dollars.'
```
